```python
# =====================================================================
# PROJECT CALLISTO: MULTILINGUAL IIT-JEE ADVANCED SPECIALIST AI
# =====================================================================
# Target Devices: Offline Android Smartphones, Tablets, and Laptops
# Supported Languages: English, Bengali, Hindi
# Core Capabilities: Advanced Physics & Mathematics Reasoning
# =====================================================================

print("Starting Project Callisto Configuration...\n")

# ---------------------------------------------------------------------
# STEP 1: INSTALLING CORE MULTIMODAL & LMM ENGINES
# ---------------------------------------------------------------------
# Installing libraries required for 4-bit quantization, LoRA adaptation,
# and high-performance inference on memory-constrained client devices.
print("Executing Step 1: Installing dependencies...")
!pip install -q unsloth transformers trl accelerate datasets tensorboard
print("Step 1 Complete: All core packages installed successfully!\n")


# ---------------------------------------------------------------------
# STEP 2: LOADING THE BASE REASONING ENGINE
# ---------------------------------------------------------------------
# Loading world's #1 open-source mathematical & scientific reasoning engine.
# Using 4-bit compression to ensure seamless offline mobile deployment.
print("Executing Step 2: Loading Base Engine from HuggingFace...")
from unsloth import FastLanguageModel
import torch

# Max sequence length set to 4096 to safely parse complex multi-step JEE Advanced proofs
max_seq_length = 4096
dtype = None
load_in_4bit = True

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/Qwen2.5-Math-7B-Instruct",
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
)
print("Step 2 Complete: 7B Base Engine loaded in 4-bit precision!\n")


# ---------------------------------------------------------------------
# STEP 3: CONFIGURE LOW-RANK ADAPTATION (LoRA) FOR CROSS-DEVICE RUNTIME
# ---------------------------------------------------------------------
# Applying Parameter-Efficient Fine-Tuning (PEFT) to optimize memory
# and lock down translation features for English, Bengali, and Hindi.
print("Executing Step 3: Initializing LoRA Layers...")
model = FastLanguageModel.get_peft_model(
    model,
    r = 16,                # High capacity rank for deeply analytical scientific parameters
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj"],
    lora_alpha = 16,
    lora_dropout = 0,
    bias = "none",
    use_gradient_checkpointing = "unsloth", # Crucial to avoid memory allocation crashes on T4
    random_state = 3407,
    use_rslora = False,
    loftq_config = None,
)
print("Step 3 Complete: Cross-device offline LoRA configuration is locked!\n")


# ---------------------------------------------------------------------
# STEP 4: ESTABLISHING THE IIT-JEE PROFESSIONAL SYSTEM PROMPT
# ---------------------------------------------------------------------
# Defining the permanent identity of the fine-tuned model as Callisto-Science.
print("Executing Step 4: Structuring Multilingual Core System Prompt...")
callisto_system_prompt = """
You are "Callisto-Science", an elite, world-class AI professor specializing strictly in IIT-JEE Advanced mathematics and physics for classes 11 and 12. Your mission is to provide highly rigorous, conceptually deep, and mathematically precise solutions to the toughest engineering entrance problems.

Core Operational Rules for Callisto-Science:
1. Identity: When asked, always identify yourself as "Callisto-Science", developed as a specialized offline reasoning AI. Never refer to yourself as Qwen or Alibaba.
2. Multilingual Multimodal Output: You must accept text/OCR problem transcripts in English, Bengali, and Hindi. No matter the input language, parse the logical steps flawlessly and output the solution in the same language requested by the user.
3. Expertise: Answer ONLY questions related to Class 11 & 12 Physics and Mathematics (IIT-JEE Main/Advanced level, Olympiads like INPhO, INMO, IPhO, IMO). If a user asks an unrelated or casual question, politely decline and redirect them to focus on Science.
4. Problem Solving Methodology:
   - Always break down complex mechanics, calculus, electrodynamics, or algebra problems into first principles (fundamental laws of physics/theorems of math).
   - Provide step-by-step mathematical derivations using clear, logical steps.
   - For physics, explicitly mention the assumptions, frame of reference, and free-body diagrams (FBD) descriptions if applicable.
5. Tone: Professional, encouraging, deeply analytical, and precise. No fluff, no robotic filler text.
"""
print("Step 4 Complete: System Prompt for IIT-JEE Advanced is fully locked and ready!\n")

print("=====================================================================")
print("PROJECT CALLISTO BASE INFRASTRUCTURE IS COMPLETELY READY FOR TRAINING!")
print("=====================================================================")
```

```python
# =====================================================================
# STEP 5: CREATING THE MULTILINGUAL IIT-JEE TRAINING DATASET
# =====================================================================
from datasets import Dataset

# Sample IIT-JEE Advanced Questions in English, Bengali, and Hindi
# This teaches Callisto how to apply its core math/physics brain across languages.
training_data = [
    {
        "instruction": "Solve the physics problem.",
        "input": "A block of mass m is placed on a smooth wedge of inclination theta. If the wedge moves horizontally with acceleration 'a' such that the block remains stationary relative to the wedge, find the value of 'a'.",
        "output": "To keep the block stationary relative to the wedge, the net force along the incline must be zero.\n1. Choose the wedge as the frame of reference (Non-inertial frame).\n2. A pseudo force of magnitude 'ma' acts horizontally opposite to the acceleration of the wedge.\n3. Resolving forces parallel to the incline:\n   - Component of gravity down the incline = mg * sin(theta)\n   - Component of pseudo force up the incline = ma * cos(theta)\n4. For relative rest: ma * cos(theta) = mg * sin(theta)\n5. Canceling mass 'm' and solving for 'a': a = g * (sin(theta) / cos(theta)) = g * tan(theta).\nConclusion: The required horizontal acceleration is a = g * tan(theta)."
    },
    {
        "instruction": "পদার্থবিদ্যার সমস্যাটি সমাধান করো।",
        "input": "একটি m ভরের বস্তু theta নতিবিশিষ্ট একটি মসৃণ নততলের (wedge) ওপর রাখা আছে। নততলটিকে অনুভূমিকভাবে কত ত্বরণ 'a' দিলে বস্তুটি নততলের সাপেক্ষে স্থির থাকবে?",
        "output": "নততলের সাপেক্ষে বস্তটিকে স্থির রাখতে হলে, নততল বরাবর প্রযুক্ত নেট বল শূন্য হতে হবে।\n১. নততলটিকে নির্দেশতন্ত্র (Non-inertial frame) হিসেবে ধরা যাক।\n২. এর ফলে বস্তুটির ওপর অনুভূমিকভাবে একটি অলীক বল (Pseudo force) 'ma' কাজ করবে, যার অভিমুখ নততলের ত্বরণের বিপরীত দিকে।\n৩. নততল বরাবর বলের উপাংশগুলি বিভাজন করলে পাই:\n   - নততল বরাবর নিচের দিকে মহাকর্ষ বলের উপাংশ = mg * sin(theta)\n   - নততল বরাবর উপরের দিকে অলীক বলের উপাংশ = ma * cos(theta)\n৪. নততলের সাপেক্ষে স্থির থাকার শর্ত: ma * cos(theta) = mg * sin(theta)\n৫. উভয় পক্ষ থেকে ভর 'm' বাদ দিলে পাই: a = g * (sin(theta) / cos(theta)) = g * tan(theta)।\nসিদ্ধান্ত: প্রয়োজনীয় অনুভূমিক ত্বরণ হলো a = g * tan(theta)।"
    },
    {
        "instruction": "गणित की समस्या को हल करें।",
        "input": "यदि log_2(x) + log_4(x) = 3 है, तो x का मान ज्ञात कीजिए।",
        "output": "दिए गए समीकरण को हल करने के लिए आधार (base) को समान करना होगा:\n1. समीकरण: log_2(x) + log_4(x) = 3\n2. हम जानते हैं कि log_4(x) = log_(2^2)(x) = (1/2) * log_2(x)\n3. इसे समीकरण में रखने पर: log_2(x) + (1/2) * log_2(x) = 3\n4. सामान्य लेने पर: (1 + 1/2) * log_2(x) = 3 => (3/2) * log_2(x) = 3\n5. दोनों तरफ 3 से भाग देने पर: (1/2) * log_2(x) = 1 => log_2(x) = 2\n6. घातीय रूप (exponential form) में बदलने पर: x = 2^2 = 4.\nनिष्कर्ष: x का मान 4 है।"
    }
]

# Formatting dataset to match Callisto's system prompt architecture
formatted_data = []
for item in training_data:
    text = f"### System:\n{callisto_system_prompt}\n\n### Instruction:\n{item['instruction']}\n\n### Input:\n{item['input']}\n\n### Response:\n{item['output']}"
    formatted_data.append({"text": text})

dataset = Dataset.from_list(formatted_data)
print("Step 5 Complete: Multilingual IIT-JEE Training Dataset successfully created and formatted!")
```

```python
# =====================================================================
# STEP 6: STARTING THE FINE-TUNING TRAINING PROCESS
# =====================================================================
from trl import SFTTrainer
from transformers import TrainingArguments
from unsloth import is_bfloat16_supported

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    dataset_text_field = "text",
    max_seq_length = max_seq_length,
    dataset_num_proc = 2,
    packing = False, # Can be True for faster training for larger datasets
    args = TrainingArguments(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 1,
        max_steps = 10, # Short steps for testing our sample dataset
        learning_rate = 2e-4,
        fp16 = not is_bfloat16_supported(),
        bf16 = is_bfloat16_supported(),
        logging_steps = 1,
        output_dir = "outputs",
    ),
)

print("Starting Model Training...")
trainer_stats = trainer.train()
print("\n=====================================================================")
print("CONGRATULATIONS OUSTAD! CALLISTO-SCIENCE TRAINING IS OFFICIALLY COMPLETE!")
print("=====================================================================")
```

```python
# Project Callisto: Safely Saving the Fine-Tuned Model locally
model.save_pretrained("callisto_jee_model")
tokenizer.save_pretrained("callisto_jee_model")
print("CONGRATULATIONS OUSTAD! Callisto-Science is officially saved successfully!")
```

```python
# =====================================================================
# FINAL STEP: LIVE TESTING YOUR CUSTOM CALLISTO-SCIENCE AI
# =====================================================================
from unsloth import FastLanguageModel

# Fast inference setup
FastLanguageModel.for_inference(model)

# Here is your official IIT-JEE Hindi Question for testing
test_question = "एक m द्रव्यमान का कण r त्रिज्या के वृत्ताकार पथ पर v एकसमान चाल से गति कर रहा है। कण पर लगने वाले अभिकेंद्रीय बल (Centripetal force) का मान और उसकी दिशा ज्ञात कीजिए।"

# Formatting the input for Callisto
inputs = tokenizer(
[
    f"### System:\n{callisto_system_prompt}\n\n### Instruction:\nSolve the science problem step-by-step.\n\n### Input:\n{test_question}\n\n### Response:\n"
], return_tensors = "pt").to("cuda")

# Generating the response using Callisto's trained brain
outputs = model.generate(**inputs, max_new_tokens = 512, use_cache = True)
response = tokenizer.batch_decode(outputs)

# Clean and print the final output
final_output = response[0].split("### Response:\n")[-1].replace("</s>", "").strip()
print("=====================================================================")
print("CALLISTO-SCIENCE LIVE RESPONSE:")
print("=====================================================================")
print(final_output)
```
